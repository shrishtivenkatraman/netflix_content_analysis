Next we analyse the top content producing companies.

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Content-Producing Countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

The United States contributes the highest number of titles, followed by India and the United Kingdom.

Now, we analyse the most frequent genres avaliable on Netflix.

```python
df["listed_in"].str.split(", ").explode().value_counts().head(10).plot(kind="bar")
plt.title("Most Common Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

Drama and international titles feature predominantly on the Netflix library, and this clearly illustrates the platform's focus on storytelling for a broad audience around the world. It appears to indicate that Netflix not only cares for popular titles in the broad genre categories but also for titles that vary in their country of origin in an effort to build a global audience.
